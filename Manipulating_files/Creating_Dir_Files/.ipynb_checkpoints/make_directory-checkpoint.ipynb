```python
# Este programa crea una serie de carpetas con archivos de nombres específicos
# Utiliza las librerias de os y una libreria propia creada para crear una lista con nombres específicos


import os
#from datetime import datetime
import files as file

mes = "Sept"
lista = file.fill_lista()


#def name_directory():
#    dia = datetime.today().strftime('%d')
#    #mes = input("¿En qué mes estamos?")
#    nombre = dia + "_" + mes + "_" + "2022"
#    return nombre

#esta función crea un directorio/folder en el path especificado y con el nombre creado
def create_directory(file_date):
    for i in range(0,31):
        parent_dir = os.getcwd()#esta funcion regresa el path del directorio actual
        new_directory = file_date
        path = os.path.join(parent_dir, file_date)#esta funcion junta el nombre de un path con otro
        if not os.path.exists(path):
            os.mkdir(path)#cread el directorio especificado
        return path

#esta función crea uno o varios archivos en el path especificado y con el nombre requerido
def create_file(file_date):
    try:
        file_ToDo = "ToDo.txt"
        file_Conclusions ="Conclusions.txt"
        pwd = os.getcwd()
        path = os.path.join(pwd, file_date)
        path_ToDo = os.path.join(path, file_ToDo)
        path_Conclusions= os.path.join(path, file_Conclusions)
        f_ToDo = open(path_ToDo,'x')
        f_Conclusions = open(path_Conclusions,'x')
    except FileExistsError:
        print("The file already exists")
    return path_ToDo, path_Conclusions

#recorre las funciones para crear la cantidad de directorios y files que uno quiera
for i in range(0,31):
    path = str(create_directory(lista[i]))
    print("Se creó el directorio: " + path)

for i in range(0,31):
    variables = create_file(lista[i])
    ToDo= variables[0]
    Conclusions = variables [1]
    print("files created:" + ToDo)
    print("files created:" + Conclusions)
```
